```python
fp = open('sales.csv')
fc = fp.read()
fp.close()
```

```python
rows = fc.split('\n')
```

```python
new_rows = []
for row in rows:
	pieces = row.split(',')
	temp = pieces[0]
	pieces[0] = pieces[1]
	pieces[1] = temp
	new_rows.append(','.join(pieces))
```

```python
print(new_rows)
```

```python
new_fc = '\n'.join(new_rows)
```

```python
new_fp = open('sales.csv','w')
new_fp.write(new_fc)
new_fp.close()
```
